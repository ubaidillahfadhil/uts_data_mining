```python
#List Library
import ydata_profiling
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
input_file = '/content/DatasetCuaca_convert.xlsx'
df = pd.read_excel(input_file)
pilih_kolom1 = ['Suhu Udara','Tekanan Udara','Kelembapan Udara','Kecepatan Angin','Curah Hujan']
pilih_kolom2 = range(1, 6)
output_file = '/content/data-cuaca-daerah_sorong.csv'
suhu_udara =['Suhu Udara']
curah_hujan =['Curah Hujan']
kelembapan_udara =['Kelembapan Udara']


def keseluruhan_data(input_file):
  display(df)
def tampilkan_pilihan(input_file, pilih_kolom1):
  selected_data = df[pilih_kolom1]
  display(selected_data)
def kecsv(input_file, output_file, pilih_kolom2):
  selected_data = df.iloc[:, pilih_kolom2]
  selected_data.to_csv(output_file, index=False)
def tampilkan_csv(output_file):
    fcsv = pd.read_csv(output_file)
    display(fcsv)

def mean_median_modus_suhu_udara(output_file, suhu_udara):
    kolom = df[suhu_udara]
    mean = kolom.mean()
    median = kolom.median()
    mode = kolom.mode().values[0]
    display(kolom)
    print("\n Berikut Nilai Mean, Median, Dan Modus Dari Suhu Udara :\n")
    print("Mean:", float(mean))
    print("Median:", float(median))
    print("Modus:", float(mode))

def mean_median_modus_curah_hujan(output_file, curah_hujan):
    kolom2 = df[curah_hujan]
    mean = kolom2.mean()
    median = kolom2.median()
    mode = kolom2.mode().values[0]
    display(kolom2)
    print("\n Berikut Nilai Mean, Median, Dan Modus Dari Curah Hujan :\n")
    print("Mean:", float(mean))
    print("Median:", float(median))
    print("Modus:", float(mode))

#4
def create_boxplot(output_file, x_column, y_column):
    df = pd.read_csv(output_file)
    plt.figure(figsize=(12, 5))
    x_column = 'Kelembapan Udara'
    y_column = 'Curah Hujan'
    sns.boxplot(x='Kelembapan Udara', y='Curah Hujan', data=df[[x_column, y_column]])
    plt.title('Boxplot Kelembapan Udara dan Curah Hujan')
    plt.xlabel('Kelembapan Udara')
    plt.ylabel('Curah Hujan')
    plt.show()

def Buat_Scatterplot(output_file, kelembapan_udara, curah_hujan):
      fcsv = pd.read_csv(output_file)
      x_data = fcsv[kelembapan_udara]
      y_data = fcsv[curah_hujan]      
      plt.figure(figsize=(12, 5))
      plt.scatter(x_data, y_data)
      plt.xlabel(kelembapan_udara)
      plt.ylabel(curah_hujan)
      plt.title("Scatter Plot Dari : {} dan {}".format(kelembapan_udara, curah_hujan))
      plt.show()
```

```python
def Deviasi_Curah_Hujan(csv_file, column):
    df = pd.read_csv(csv_file)
    data = df[column]
    std_dev = data.std()
    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=10, alpha=0.5, color='blue', edgecolor='black', label=column)
    plt.axvline(x=std_dev, color='orange', linestyle='--', linewidth=2, label='Standard Deviation')
    plt.title('Histogram Dari: {}'.format(column))
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
def Deviasi_Suhu_Udara(csv_file, column2):
    df = pd.read_csv(csv_file)
    data = df[column2]
    std_dev = data.std()
    plt.figure(figsize=(8, 6))
    plt.hist(data, bins=10, alpha=0.5, color='blue', edgecolor='black', label=column2)
    plt.axvline(x=std_dev, color='orange', linestyle='--', linewidth=2, label='Standard Deviation')
    plt.title('Histogram Dari: {}'.format(column2))
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()    
csv_file = output_file
column = 'Curah Hujan'
column2 = 'Suhu Udara'
Deviasi_Curah_Hujan(csv_file, column)
Deviasi_Suhu_Udara(csv_file, column2)
```

```python
# Membaca dataset
data = pd.read_csv('/content/data-cuaca-daerah_sorong.csv')

# Menampilkan informasi umum tentang dataset
data_info = data.describe(include='all').transpose()

# Menghitung jumlah missing value pada setiap kolom
missing_values = data.isnull().sum()

# Menghitung korelasi antar kolom
correlation = data.corr()

# Menampilkan statistik kuantitatif untuk kolom numerik
numeric_stats = data.select_dtypes(include=['float64', 'int64']).describe().transpose()

# Menampilkan mode untuk kolom kategorikal
categorical_mode = data.select_dtypes(include=['object']).mode().transpose()

# Mengidentifikasi outlier menggunakan metode IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
outliers = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).sum()

# Menyusun laporan non-grafis
report = pd.concat([data_info, missing_values.rename('Missing Values'), numeric_stats, categorical_mode.rename(columns={0: 'Mode'}), outliers.rename('Outliers')], axis=1)

# Menampilkan laporan non-grafis
print(report)
```
